```python
from sklearn import preprocessing

df1.convert_objects(convert_numeric=True)
df1.fillna(0, inplace=True)

def textToNumbericConverter(df1):
    columns = df1.columns.values

    for col in columns:
        tex_digit_vals = {}
        def convertToInt(val):
            return tex_digit_vals[val]
        if df1[col].dtype != np.int64 and df1[col].dtype != np.float64:
            column_contents = df1[col].values.tolist()
            unique_elements = set(column_contents)
            x=0
            for unique in unique_elements:
                if unique not in tex_digit_vals:
                    tex_digit_vals[unique] = x
                    x +=1
            df1[col] = list(map(convertToInt, df1[col]))
    return df1
df1 = textToNumbericConverter(df1)
df1
```
